```python
from kiwoom.kiwoom import *
```

```python
save_list = []
```

```python
kiwoom = Kiwoom(save_list)
```

```python
len(kiwoom.calculator_list)
```

```python
stock_code_list = []
stock_name_list = []
price_volume_list = []
```

```python
for ci in kiwoom.calculator_list:
    stock_code_list.append(ci[1])
    stock_name_list.append(ci[2])
    price_volume_list.append(ci[4])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = stock_name_list, index = ['buy_pricevolume'])
```

```python
for si, stock_name in enumerate(stock_name_list):
    df[stock_name] = [price_volume_list[si]]
```

```python
df
```

```python
df.to_csv('weekly_top_buy_list.csv', index=True)
```

```python
stock_name_list
```

```python
closeprice_shareoutstanding_df = pd.DataFrame(columns = stock_code_list, index = ['closeprice', 'shareoutstanding']) 
for idx, stock_code in enumerate(stock_code_list):
    stockwise_result_list = []
    stockwise_result_list.append(int(kiwoom.dynamicCall("GetMasterLastPrice(QString)", stock_code)))
    stockwise_result_list.append(int(kiwoom.dynamicCall("GetMasterListedStockCnt(QString)", stock_code)))
    closeprice_shareoutstanding_df[stock_code] = stockwise_result_list
```

```python
closeprice_shareoutstanding_df
```

```python
pension_fund_df = df
```

```python
pension_fund_df
```

```python
closeprice_shareoutstanding_df
```

```python
closeprice_shareoutstanding_df.columns = pension_fund_df.columns
```

```python
total_pension_fund_df= pd.concat([pension_fund_df, closeprice_shareoutstanding_df])
```

```python
total_pension_fund_df
```

```python
def parse_to_int(string_number_with_sign):
    return int(string_number_with_sign)
```

```python
total_pension_fund_df.loc['pension_fund_net']=total_pension_fund_df.loc['buy_pricevolume'].apply(lambda x: parse_to_int(x))
```

```python
total_pension_fund_df
```

```python
total_pension_fund_df.loc['pension_fund_net_mil']= total_pension_fund_df.loc['pension_fund_net']*1000000
```

```python
total_pension_fund_df.loc['market_cap']= total_pension_fund_df.loc['closeprice']*total_pension_fund_df.loc['shareoutstanding']
```

```python
total_pension_fund_df
```

```python
total_pension_fund_df.loc['pension_portion']=total_pension_fund_df.loc['pension_fund_net_mil']/total_pension_fund_df.loc['market_cap']
```

```python
total_pension_fund_df
```

```python
pension_portion_df = total_pension_fund_df.transpose()[['pension_portion']]
```

```python
pension_portion_df
```

```python
rank_pension_portion_df= pension_portion_df.sort_values(by='pension_portion', ascending=False)
```

```python
rank_stock_code_list = list(rank_pension_portion_df.index)
```

```python
rank_stock_name_list = [kiwoom.dynamicCall("GetMasterCodeName(QString)", code) for code in rank_stock_code_list]
```

```python
rank_pension_portion_df['stock_name']=rank_stock_name_list
```

```python
rank_pension_portion_df.head(50)
```

```python
pension_amount_df = total_pension_fund_df.transpose()[['pension_fund_net_mil']]
```

```python
rank_pension_amount_df = pension_amount_df.sort_values(by='pension_fund_net_mil', ascending=False)
```

```python
rank_stock_code_list = list(rank_pension_amount_df.index)
```

```python
rank_stock_name_list = [kiwoom.dynamicCall("GetMasterCodeName(QString)", code) for code in rank_stock_code_list]
```

```python
rank_pension_amount_df['stock_name']=rank_stock_name_list
```

```python
rank_pension_amount_df.head(50)
```
