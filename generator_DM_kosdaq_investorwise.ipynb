```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
import datetime
```

```python
target_year = '2021'
```

```python
full_date_df = fdr.DataReader('KQ11', target_year)
```

```python
full_date_df
```

```python
raw_tradingdays = list(full_date_df.index.values)
```

```python
tradingdays = [int(pd.to_datetime(td).strftime('%Y%m%d')) for td in raw_tradingdays]
```

```python
daily_folder_dir = './daily_stockwise_dict/'+target_year
```

```python
import sys

from PyQt5.QtWidgets import *

app = QApplication(sys.argv)

from kiwoom.kiwoom import *

save_list = []

kiwoom = Kiwoom(save_list)
```

```python
stock_code_list = kiwoom.get_code_list_by_market('10') # '10'이 코스닥
```

```python
stock_code_list
```

```python
stockwise_list_dict = {}
for stock_code in stock_code_list:
    stockwise_list_dict[stock_code] = []
```

```python
import pandas as pd
```

```python
import pickle
```

```python
# 여기서 일자별로 순회
```

```python
# 외국인: 1
# 기관: 2
# 연기금: 8
```

```python
for curr_tradingday in tradingdays:
    daily_file_1_dir = daily_folder_dir+'/daily_stockwise_dict_'+str(curr_tradingday)+'_1.pickle'
    daily_file_2_dir = daily_folder_dir+'/daily_stockwise_dict_'+str(curr_tradingday)+'_2.pickle'
    with open(daily_file_1_dir, 'rb') as fr:
        stockwise_dict_1 = pickle.load(fr)
    with open(daily_file_2_dir, 'rb') as fr:
        stockwise_dict_2 = pickle.load(fr)
    for stock_code in stock_code_list:
        if stock_code in list(stockwise_dict_1.keys()):
            stockwise_list_dict[stock_code].append(stockwise_dict_1[stock_code][2])
        elif stock_code in list(stockwise_dict_2.keys()):
            stockwise_list_dict[stock_code].append(stockwise_dict_2[stock_code][2])
        else:
            stockwise_list_dict[stock_code].append(0)
```

```python
tradingdays.index(20220103)
```

```python
year_daily_df = pd.DataFrame(columns = stock_code_list, index = tradingdays[:248])
```

```python
for stock_code in stock_code_list:
    year_daily_df[stock_code] = stockwise_list_dict[stock_code]
```
