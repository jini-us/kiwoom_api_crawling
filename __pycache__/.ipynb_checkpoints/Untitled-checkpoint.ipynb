```python
sum_start_date = 20210711
sum_end_date = 20210810
```

```python
target_tradingdays = [td for td in tradingdays if td >= sum_start_date and td<=sum_end_date]
```

```python
# 연기금 매수 데이터 누적
pension_fund_buy_dict = {}
for target_tradingday in target_tradingdays:
    target_start_date = target_tradingday
    target_end_date = target_tradingday
    target_buy_or_sale = 2
    target_market = 101
    target_investor = 6000
    kiwoom.calculator_list = []
    kiwoom.get_daily_buy_or_sale_of_investor(start_date=target_start_date, end_date=target_end_date, buy_or_sale=target_buy_or_sale, market_code=target_market, investor_code=target_investor)
    daily_result = kiwoom.calculator_list
    pension_fund_buy_dict[target_tradingday] = daily_result
```
