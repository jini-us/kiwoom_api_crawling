```python
return_df = ((close_price_df - close_price_df.shift(1))/close_price_df.shift(1))*100
```

```python
tradingday_return_dict = {}
for day in tradingdays[3:-1]:
    tradingday_return_dict[day]=[]
    daily_institution_buy_3dr_df = institution_buy_3dr_df.loc[day:day]
    daily_pension_buy_3dr_df = pension_buy_3dr_df.loc[day:day]
    institution_sorted_amount = sorted(daily_institution_buy_3dr_df.values.tolist()[0], reverse=True)
    institution_topten_amount = institution_sorted_amount[5:20]
    pension_sorted_amount = sorted(daily_pension_buy_3dr_df.values.tolist()[0], reverse=True)
    pension_topten_amount = pension_sorted_amount[5:20]
    for stock_code in stock_code_list:
        daily_return_df = return_df.loc[day:day]
        if daily_return_df[stock_code].values[0] > 0:
            continue
        if (daily_institution_buy_3dr_df[stock_code].values[0] in institution_topten_amount) and (daily_pension_buy_3dr_df[stock_code].values[0] in pension_topten_amount):
            tradingday_return_dict[day].append(stock_code)
```

```python
tradingday_return_dict
```

```python
# 단순히 당일 오르고 내림을 계산해보자
up_cnt = 0
down_cnt = 0
for tradingday in tradingday_return_dict.keys():
    daily_top_amount_stock_code = tradingday_return_dict[tradingday]
    for stock_code in daily_top_amount_stock_code:
        daily_stock_return = return_df[stock_code][tradingday]
        if np.isnan(daily_stock_return):
            continue
        elif daily_stock_return > 0:
            up_cnt += 1
        elif daily_stock_return < 0:
            down_cnt += 1
        else:
            continue
```

```python
print(up_cnt)
print(down_cnt)
```

```python
# 단순히 다음날 오르고 내림을 계산해보자
up_cnt = 0
down_cnt = 0
for tradingday in tradingday_return_dict.keys():
    daily_top_amount_stock_code = tradingday_return_dict[tradingday]
    for stock_code in daily_top_amount_stock_code:
        tomorrow_index = tradingdays.index(tradingday)+3
        if tomorrow_index >=len(tradingdays)-1:
            continue
        daily_stock_return = return_df[stock_code][tradingdays[tomorrow_index]]
        if np.isnan(daily_stock_return):
            continue
        elif daily_stock_return > 0:
            up_cnt += 1
        elif daily_stock_return < 0:
            down_cnt += 1
        else:
            continue
```
