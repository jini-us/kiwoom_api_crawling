```python
import sys
from PyQt5.QtWidgets import *
from kiwoom.kiwoom_auto import *
import pickle
import os
import pandas as pd
```

```python
class DM_loader():
    
    def __init__(self):
        self.DM_dir = 'C:\\Users\\박형진\\Desktop\\pythonTrading\\DM'
        self.app = QApplication(sys.argv)
        self.kiwoom = Kiwoom([])
        self.kosdaq_stock_code_list = self.kiwoom.get_code_list_by_market('10')
    
    def get_market_stock_list(self, target_market):
        if target_market=='kosdaq':
            return self.kosdaq_stock_code_list
    
    def get_data(self, target_market, target_buyer, start_yearmonth, end_yearmonth):
        
        
        target_data_dir = os.path.join(self.DM_dir, ('DM_'+target_market+'_'+target_buyer+'_buy'))
        total_yearmonth_list = [filename.split('.')[0] for filename in os.listdir(target_data_dir)]
        result_yearmonth_list = [yearmonth for yearmonth in total_yearmonth_list if (yearmonth>=start_yearmonth and yearmonth<=end_yearmonth)]
        
        result_df = pd.DataFrame(columns = self.get_market_stock_list(target_market))
        for result_yearmonth in result_yearmonth_list:
            yearmonth_df_dir = os.path.join(target_data_dir, (result_yearmonth+'.pickle'))
            with open(yearmonth_df_dir, 'rb') as fr:
                yearmonth_df = pickle.load(fr)
            
            yearmonth_df = yearmonth_df.reindex(columns = result_df.columns)
            result_df = pd.concat([result_df, yearmonth_df])
        return result_df
```

```python
loader = DM_loader()
```

```python
pension_buy_df = loader.get_data('kosdaq', 'pension', '202101', '202109')
institution_buy_df = loader.get_data('kosdaq', 'institution', '202101', '202109')
foreigner_buy_df = loader.get_data('kosdaq', 'foreigner', '202101', '202109')
```

```python
pension_buy_df
```

```python
pension_shift_df = pension_buy_df.shift(1)
institution_shift_df = institution_buy_df.shift(1)
foreigner_shift_df = foreigner_buy_df.shift(1)
```

```python
import numpy as np
```

```python
pension_changed_m = np.bitwise_and(pension_buy_df.values>0, pension_shift_df.values<0)
institution_changed_m = np.bitwise_and(institution_buy_df.values>0, institution_shift_df.values<0)
foreigner_changed_m = np.bitwise_and(foreigner_buy_df.values>0, foreigner_shift_df.values<0)
```

```python
all_changed_m = np.bitwise_and(pension_changed_m, institution_changed_m)
all_changed_m = np.bitwise_and(all_changed_m, foreigner_changed_m)
```
