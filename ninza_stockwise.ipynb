```python
sum_start_date = 20210712
sum_end_date = 20210811
```

```python
stockwise_dict = {}
```

```python
target_start_date = sum_start_date
target_end_date = sum_end_date
target_price_or_volume = 1 # 금액: 1, 수량: 2
target_net_or_buy_or_sale = 0 # 순매수: 0, 매수: 1, 매도: 2
target_unit_code = 1 # 0: 천 주, 1: 단 주
```

```python
stock_code_list = kiwoom.get_code_list_by_market('10') # '10'이 코스닥
```

```python
stock_code_list
```

```python
for idx, target_stock_code in enumerate(stock_code_list):
    if idx == 800:
        break
    print(str(idx) + '/' + str(len(stock_code_list)) + ' 현재 ' + str(target_stock_code) + '의 매매 데이터를 불러오는 중')
    kiwoom.calculator_list = []
    kiwoom.get_cummulative_buy_or_sale_of_investor_stockwise(stock_code=target_stock_code, start_date = target_start_date, end_date = target_end_date, price_or_volume = target_price_or_volume, net_or_buy_or_sale = target_net_or_buy_or_sale, unit_code = target_unit_code)
    raw_data = kiwoom.calculator_list[0]
    parsed_data = [int(rd.strip()) for rd in raw_data]
    stockwise_dict[target_stock_code] = parsed_data
```

```python
pension_fund_df = pd.DataFrame(columns=stock_code_list, index=['pension_fund_net'])

for idx, stock_code in enumerate(stock_code_list):
    pension_fund_df[stock_code] = [stockwise_dict[stock_code][8]]
```

```python
pension_fund_df
```

```python
institution_df = pd.DataFrame(columns=stock_code_list, index=['institution_net'])

for idx, stock_code in enumerate(stock_code_list):
    institution_df[stock_code] = [stockwise_dict[stock_code][2]]
```

```python
institution_df
```

```python
closeprice_shareoutstanding_df = pd.DataFrame(columns = stock_code_list, index = ['closeprice', 'shareoutstanding']) 
for idx, stock_code in enumerate(stock_code_list):
    stockwise_result_list = []
    stockwise_result_list.append(int(kiwoom.dynamicCall("GetMasterLastPrice(QString)", stock_code)))
    stockwise_result_list.append(int(kiwoom.dynamicCall("GetMasterListedStockCnt(QString)", stock_code)))
    closeprice_shareoutstanding_df[stock_code] = stockwise_result_list
```
